```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import scipy.stats as stats
```

```python
df_alt_vs_prec = pd.read_csv("Precipitation2018.csv", index_col=0)

df_alt_vs_prec.head()
```

```python
X=pd.DataFrame(df_alt_vs_prec.altitude)
Y=pd.DataFrame(df_alt_vs_prec.t_total_prec_value)
x=pd.DataFrame.to_numpy(X)
y=pd.DataFrame.to_numpy(Y)


x=x.reshape(-1,1)
linreg= LinearRegression()
linreg.fit(x,y)
y_pred=linreg.predict(x)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x[:,0],y[:,0])
r_squared=round(np.square(r_value),2)
p_value=round(p_value,2)
std_err=round(std_err,2)
fig, ax = plt.subplots(figsize=(20,10))
#plt.show()

#plot the points
ax.scatter(df_alt_vs_prec.altitude,df_alt_vs_prec.t_total_prec_value,s=250,edgecolors='black', c="green")
ax.set_xlabel('ALTITUDE',size=20)
ax.set_ylabel('PRECIPITATION',size=20)
plt.xticks(fontsize= 20)
plt.yticks(fontsize= 20)
ax.set_title(('ALTITUDE vs PRECIPITATION for stations located in the 13 counties '),size=30,pad=30)

plt.plot(x, y_pred, color='red')
plt.text(0.75, 0.75, 'r_squared = ' + str(r_squared) + '\n' + 'standard error =' + str(std_err) + '\n' + 'p_value =' + str(p_value), horizontalalignment='left',verticalalignment='top', transform=ax.transAxes,bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=1'),size=20)

plt.savefig("Ex5-ALTITUDEvsPRECIPITATION.png", bbox_inches='tight')
```
