```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import scipy.stats as stats
```

```python
filepathname = r"../Exercise 7/NDVI_vs_SMI.csv"
df_NDVI_vs_SMI = pd.read_csv(filepathname)
```

```python
df_NDVI_vs_SMI
```

```python
X=pd.DataFrame(df_NDVI_vs_SMI.SMI_all)
Y=pd.DataFrame(df_NDVI_vs_SMI.NDVI_all)
x=pd.DataFrame.to_numpy(X)
y=pd.DataFrame.to_numpy(Y)


x=x.reshape(-1,1)
linreg= LinearRegression()
linreg.fit(x,y)
y_pred=linreg.predict(x)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x[:,0],y[:,0])
r_squared=round(np.square(r_value),2)
p_value=round(p_value,2)
std_err=round(std_err,2)
fig, ax = plt.subplots(figsize=(20,10))
#plt.show()

#plot the points
ax.scatter(df_NDVI_vs_SMI.SMI_all,df_NDVI_vs_SMI.NDVI_all,s=150,edgecolors='black')
ax.set_xlabel('SMI',size=30)
ax.set_ylabel('NDVI',size=30)
plt.xticks(fontsize= 20)
plt.yticks(fontsize= 20)
ax.set_title(('SMI vs NDVI for Region of Interest (ROI) '),size=40,pad=20)

plt.plot(x, y_pred, color='red')
plt.text(0.75, 0.75, 'r_squared = ' + str(r_squared) + '\n' + 'standard error =' + str(std_err) + '\n' + 'p_value =' + str(p_value), horizontalalignment='left',verticalalignment='top', transform=ax.transAxes,bbox=dict(facecolor='none', edgecolor='black', boxstyle='round,pad=1'),size=20)

plt.savefig("../Exercise 7/Exc7_smi_vs_ndvi.png", bbox_inches='tight')
```
