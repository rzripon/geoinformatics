```python
# Import necessary packages
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import seaborn as sns
import pandas as pd


# Handle date time conversions between pandas and matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Use white grid plot background from seaborn
sns.set(font_scale=1.5, style="whitegrid")
```

```python
# Import data using datetime and no data value
precip_2018_daily = pd.read_csv('PrecipitationTotalCumulative.csv',
                                     parse_dates=['date'],
                                     index_col= ['date'],
                                     na_values=['999.99'])
```

```python
precip_2018_daily.head(10)
```

```python
precipation_2018 = precip_2018_daily['2018-04-16':'2018-08-16']

precipation_2018.head()
```

```python
# Create figure and plot space
fig, ax = plt.subplots(figsize=(12, 4))

# Add x-axis and y-axis
ax.bar(precipation_2018.index.values,
       precipation_2018['station_id_216'],
       color='green')

plt.xticks(rotation=45)
# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Precipitation Value",
       title="Station 216 (OE)")

# Define the date format
date_form = DateFormatter("%d-%m-%y")
ax.xaxis.set_major_formatter(date_form)

# Ensure a major tick for each week using (interval=1) 
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
plt.savefig("TotalPrecipitationStation216.png", bbox_inches='tight')
plt.show()
```
